```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('emma_salt_pepper.png', cv2.IMREAD_GRAYSCALE)

# (a) Apply Gaussian Smoothing
gaussian_blur = cv2.GaussianBlur(img, (5, 5), 1.5)
cv2.imwrite('results/ques8/gaussian_blur_emma_salt_pepper.png', gaussian_blur)

# (b) Apply Median Filtering
median_blur = cv2.medianBlur(img, 5)
cv2.imwrite('results/ques8/median_blur_emma_salt_pepper.png', median_blur)

# --- Visualization ---
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original (Salt & Pepper)")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gaussian_blur, cmap='gray')
plt.title("Gaussian Smoothing")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(median_blur, cmap='gray')
plt.title("Median Filtering")
plt.axis('off')

plt.tight_layout()
plt.show()
```
